```python
# Script: verify_rebuilt_datasets.py
import numpy as np
import os
import matplotlib.pyplot as plt

def verify_dataset(set_name):
    # Cargar datos
    X = np.load(f'../Datasets_DeepShadows/array_images/X_{set_name}.npy')
    y = np.load(f'../Datasets_DeepShadows/Galaxies_data/y_{set_name}.npy')
    
    # Verificar tamaos
    assert len(X) == len(y), f"Error de tamao! X: {len(X)}, y: {len(y)}"
    
    # Mostrar muestras aleatorias
    plt.figure(figsize=(15, 10))
    for i in range(10):
        idx = np.random.randint(len(X))
        plt.subplot(2, 5, i+1)
        plt.imshow(X[idx])
        plt.title(f"Label: {y[idx]}")
        plt.axis('off')
    plt.suptitle(f"Muestras de {set_name} (n={len(X)})", fontsize=16)
    plt.show()
    
    # Estadsticas de etiquetas
    galaxy_count = np.sum(y == 1)
    print(f"\nConjunto {set_name}:")
    print(f"Total muestras: {len(X)}")
    print(f"Galaxias: {galaxy_count} ({galaxy_count/len(X):.2%})")
    print(f"Artefactos: {len(X) - galaxy_count}")

# Verificar todos los conjuntos
for dataset in ['train', 'val', 'test']:
    verify_dataset(dataset)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_samples(set_name, num_samples=10):
    # Cargar datos
    X = np.load(f'../Datasets_DeepShadows/array_images/X_{set_name}.npy')
    y = np.load(f'../Datasets_DeepShadows/Galaxies_data/y_{set_name}.npy')
    
    # Seleccionar muestras aleatorias
    indices = np.random.choice(len(X), num_samples, replace=False)
    
    # Configurar figura
    plt.figure(figsize=(15, 8))
    plt.suptitle(f'Muestras del conjunto {set_name} (n={len(X)})', fontsize=16)
    
    for i, idx in enumerate(indices):
        plt.subplot(2, 5, i+1)
        plt.imshow(X[idx])
        plt.title(f"Label: {y[idx]}\n{'Galaxia' if y[idx]==1 else 'Artefacto'}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # Estadsticas
    galaxy_count = np.sum(y == 1)
    print(f"\nConjunto {set_name}:")
    print(f"Total muestras: {len(X)}")
    print(f"Galaxias: {galaxy_count} ({galaxy_count/len(X):.2%})")
    print(f"Artefactos: {len(X) - galaxy_count}")

# Visualizar todos los conjuntos
for dataset in ['train', 'val', 'test']:
    visualize_samples(dataset)
```
