```python
# Import basic packages
import numpy as np
import scipy as sp
import pandas as pd
from scipy import interpolate
import matplotlib.pyplot as plt
import random
```

```python
# For image-based classification
X_tr = np.float64(np.load("../Data_training/class_1_combined.npy")) #Training
```

```python
# Define the grid dimensions (n x m) for plotting
n_rows = 8  # Number of rows
m_cols = 5  # Number of columns
```

```python
# Create a grid for plotting images
fig, axes = plt.subplots(n_rows, m_cols, figsize=(12, 12))

for i in range(n_rows):
    for j in range(m_cols):
        index = i * m_cols + j
        if index < len(X_tr):
            axes[i, j].imshow(X_tr[index])
            axes[i, j].axis('off')

# Adjust layout and display the grid of images
plt.tight_layout()
plt.show()
```
